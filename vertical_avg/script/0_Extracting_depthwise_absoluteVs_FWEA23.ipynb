```python
from netCDF4 import Dataset
import numpy as np
import os

dir=f"/media/rupak/Rupak_4TB/BACKUPS/MANUSCRIPTS_Rupak/SE_ASIA_MANU_DRAFTS/MERGED/Manu_figs_organised/text_fig/GitHub_upload/model" #path to the model nc file

filen=os.path.join(dir,f'FWEA23.r0.0-n4.nc')

nc_f=Dataset(filen,'r')
dp=nc_f.variables['depth'][:]
lat=nc_f.variables['latitude'][:]
lon=nc_f.variables['longitude'][:]
vsv=nc_f.variables['vsv'][:]
vsh=nc_f.variables['vsh'][:]
beta=nc_f.variables['beta'][:]

#puttinh in arrays
dp_array=np.array(dp)
lat_array=np.array(lat)
lon_array=np.array(lon)
vsv_array=np.array(vsv)
vsh_array=np.array(vsh)
beta_array=np.array(beta)

#print(f"depth={dp_array}")
#print(f"{dp_array[60]}")

#shape
print(f"dp shape:{dp_array.shape}")
print(f"lat_shape:{lat_array.shape}")
print(f"lon_shape:{lon_array.shape}")
print(f"vsv_shape:{vsv_array.shape}")
print(f"vsh_shape:{vsh_array.shape}")
print(f"beta_shape{beta_array.shape}")

#defining depths for which I want information
#des_dep=[650]
des_dep=[100,130,150,170,200,220]
#finding index of depth
for i in des_dep:
    dp_index=np.where(dp_array==i)[0]
    print(f"index of {i} km is {dp_index}")
    
    if dp_index.size == 0:
        print(f"Depth {i} km not found in the dataset.")
        continue
    
    dp_index = dp_index[0]  # Extract the integer index
    
    vsv_at_dp=vsv_array[dp_index,:,:]
    vsh_at_dp=vsh_array[dp_index,:,:]
    beta_at_dp=beta_array[dp_index,:,:]
    
    #print(f"vsv at {i} km depth is {vsv_at_dp}")
    odir=f"/media/rupak/Rupak_4TB/BACKUPS/MANUSCRIPTS_Rupak/SE_ASIA_MANU_DRAFTS/MERGED/Manu_figs_organised/text_fig/GitHub_upload/vertical_avg/files"  #output directory
    
    out=os.path.join(odir,f'FWEA23_lon_lat_Vs_{i}_final.txt')
    with open(out,'w') as f:
    
        for j in range(len(lat_array)):
            for k in range(len(lon_array)):
                #print(f"dp:{i} lon:{lon[k]} lat:{lat[j]} vsv_at_dp:{vsv_at_dp[j,k]} ")
            
                #saving avg Vs (Voigt avg)
                vsv_sqr=np.square(vsv_at_dp[j,k])
                vsh_sqr=np.square(vsh_at_dp[j,k])
                A=2*vsv_sqr
                B=(A+vsh_sqr)/3
                Vs=np.sqrt(B)
            
                #print(f"lon:{lon_array[k]} lat:{lat_array[j]} Vs:{Vs}")
            
                f.write(f"{lon_array[k]} {lat_array[j]} {Vs}\n")
            
    print(f"Extracting FWEA23 model prm for {i} km depth is done")
    
    
    print(f"Output -- {out}")
    print("")
            


```
