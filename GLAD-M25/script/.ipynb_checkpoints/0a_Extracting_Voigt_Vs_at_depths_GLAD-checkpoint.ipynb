# Extract infromation from the nc file

```python
#This code extracts Voigt Vs values from vsh and vsv for user fed depths from GLAD-M25
import os
from netCDF4 import Dataset
import numpy as np

dpath=f'/media/rupak/Rupak_4TB/BACKUPS/MANUSCRIPTS_Rupak/SE_ASIA_MANU_DRAFTS/MERGED/Manu_figs_organised/text_fig/GitHub_upload/GLAD-M25/GLAD_model'

ncf=os.path.join(dpath,f'glad-m25-vs-0.0-n4.nc')
print(f"Reading file:")
print(f"{ncf}")

ncf=Dataset(ncf,'r')
dp=ncf.variables['depth'][:]
lon=ncf.variables['longitude'][:]
lat=ncf.variables['latitude'][:]
vsh=ncf.variables['vsh'][:]
vsv=ncf.variables['vsv'][:]

#array
lon_array=np.array(lon)
lat_array=np.array(lat)
dp_array=np.array(dp)
vsh_array=np.array(vsh)
vsv_array=np.array(vsv)

#shape
print(f"Shape of lon_array:{lon_array.shape}")
print(f"Shape of lat array: {lat_array.shape}")
print(f"Shape of dp array: {dp_array.shape}")
print(f"Shape of vsh array:{vsh_array.shape}") #(dp,lat,lon)
print(f"Shape of vsv array: {vsv_array.shape}")

#desired depth from user
des_dp=[100,110,120,130,140,150,160,170,180,190,200,220]

for i in des_dp:
    print(f"Desired depth : {i} km")
    dp_idx=np.where(dp_array == i)[0]
    dp_index=dp_idx[0] #Extracting the interger
    print(f"The index for {i} km depth in depth array is {dp_index}")
    print(f"Making sure...")
    print(f"The depth value at {dp_index} th index position is {dp_array[dp_index]}")
    #print(f"")
    
    vsv_at_dp=vsv_array[dp_index,:,:]
    vsh_at_dp=vsh_array[dp_index,:,:]
    
    #output
    odpath=
    out=os.path.join(dpath,f"GLAD_lon_lat_VoigtVs_{i}km.txt")
    with open(out,'w') as fi:
    
        for j in range(len(lat_array)):
            for k in range(len(lon_array)):
                #print(f"lon:{lon_array[k]} lat:{lat_array[j]} vsv:{vsv_at_dp[j,k]} vsh:{vsh_at_dp[j,k]}")
            
                vsv_sqr=np.square(vsv_at_dp[j,k])
                vsh_sqr=np.square(vsh_at_dp[j,k])
                A=2*vsv_sqr
                B=(A+vsh_sqr)/3
                Vs=np.sqrt(B)
            
                #print(f"{lon_array[k]} {lat_array[j]} {Vs}")
                fi.write(f"{lon_array[k]} {lat_array[j]} {Vs}\n")
            
    print(f"Output-{out}")
    print(f"")
          
      
```
