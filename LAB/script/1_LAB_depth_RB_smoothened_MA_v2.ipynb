# USe this Notebook for LAB calculation. This first applies a moving avg filter to the VSH along depth directions and then
%finds the VSH_min from 40-300 km interval
%and then searhes for the gradient change point from 0-300 km interval
%and then the mean of these 2 depth is the lab depth corresponding to that node

```python
##here I am finding the VSh min within 40-300 km , but then search the smallest VSh to 0-300 km.

import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import os

# Load the NetCDF data
ncf = f"/media/rupak/Rupak_4TB/BACKUPS/MANUSCRIPTS_Rupak/SE_ASIA_MANU_DRAFTS/MERGED/Manu_figs_organised/text_fig/GitHub_upload/model/FWEA23.r0.0-n4.nc"
nf = Dataset(ncf, 'r')
dp = nf.variables['depth'][:]
lon = nf.variables['longitude'][:]
lat = nf.variables['latitude'][:]
vsh = nf.variables['vsh'][:]

# Arrays
dp_array = np.array(dp)
lon_array = np.array(lon)
lat_array = np.array(lat)
vsh_array = np.array(vsh)

# Smooth the Vsh data
def apply_moving_average_vsh(vsh_data, win):
    dp_dim, lat_dim, lon_dim = vsh_data.shape
    smoothed_vsh = np.empty((dp_dim - win + 1, lat_dim, lon_dim))

    for i in range(vsh.shape[1]):
        for j in range(vsh.shape[2]):
            depth_profile = vsh_data[:, i, j]
            smoothed_profile = moving_average(depth_profile, win)
            smoothed_vsh[:, i, j] = smoothed_profile

    return smoothed_vsh

def moving_average(vsh, win):
    return np.convolve(vsh, np.ones(win) / win, mode='valid')

# Apply smoothing
win = 7  # Window size for smoothing
smoothed_vsh_array = apply_moving_average_vsh(vsh_array, win)

# Define depth ranges
dp_min_vsh_range = 40  # Minimum depth for finding Vsh_min
dp_max_vsh_range = 300  # Maximum depth for finding Vsh_min
dp_min_grad_range = 0  # Minimum depth for gradient calculation
dp_max_grad_range = 300  # Maximum depth for gradient calculation

# Extract depth indices within each range
dp_indices_vsh = np.where((dp_array >= dp_min_vsh_range) & (dp_array <= dp_max_vsh_range))[0]
dp_indices_grad = np.where((dp_array >= dp_min_grad_range) & (dp_array <= dp_max_grad_range))[0]

# Extract smoothened Vsh within the defined depth ranges
smooth_vsh_subset_vsh_range = smoothed_vsh_array[dp_indices_vsh, :, :]
smooth_vsh_subset_grad_range = smoothed_vsh_array[dp_indices_grad, :, :]

# Extracting within the specified lat-long
lon_indices = np.where((lon_array >= 90) & (lon_array <= 120))[0]
lat_indices = np.where((lat_array >= 5) & (lat_array <= 30))[0]
smooth_vsh_subset_vsh_range_lat_lon = smooth_vsh_subset_vsh_range[:, lat_indices][:, :, lon_indices]
smooth_vsh_subset_grad_range_lat_lon = smooth_vsh_subset_grad_range[:, lat_indices][:, :, lon_indices]

# LAB calculation from the smoothened data within Vsh range (40-300 km)
min_smoothened_Vsh = np.empty((smooth_vsh_subset_vsh_range_lat_lon.shape[1], smooth_vsh_subset_vsh_range_lat_lon.shape[2]))
depth_of_min_vsh = np.empty((smooth_vsh_subset_vsh_range_lat_lon.shape[1], smooth_vsh_subset_vsh_range_lat_lon.shape[2]))

# Finding the minimum Vsh in the 40-300 km range
for i in range(smooth_vsh_subset_vsh_range_lat_lon.shape[1]):
    for j in range(smooth_vsh_subset_vsh_range_lat_lon.shape[2]):
        min_vsh = np.min(smooth_vsh_subset_vsh_range_lat_lon[:, i, j])
        min_vsh_index = np.where(smooth_vsh_subset_vsh_range_lat_lon[:, i, j] == min_vsh)[0][0]
        min_smoothened_Vsh[i, j] = min_vsh
        depth_of_min_vsh[i, j] = dp_array[dp_indices_vsh[min_vsh_index]]

# Initialize upward_depths with NaN
upward_depths = np.full((smooth_vsh_subset_vsh_range_lat_lon.shape[1], smooth_vsh_subset_vsh_range_lat_lon.shape[2]), np.nan)

# Calculate the gradient and find upward depth in the 0-300 km range
for i in range(smooth_vsh_subset_vsh_range_lat_lon.shape[1]):
    for j in range(smooth_vsh_subset_vsh_range_lat_lon.shape[2]):
        # Get the minimum Vsh depth index
        start_depth_index = np.where(dp_array == depth_of_min_vsh[i, j])[0][0]
        if start_depth_index < smooth_vsh_subset_grad_range_lat_lon.shape[0]:
            # Extract the profile for gradient calculation within 0-300 km
            vsh_profile_grad = smooth_vsh_subset_grad_range_lat_lon[:, i, j]
            depths_profile_grad = dp_array[dp_indices_grad]

            # Calculate the gradient for the entire 0-300 km profile
            gradient = np.gradient(vsh_profile_grad, depths_profile_grad)
            
            # Move upwards in depth and find where gradient is neutral or positive above min Vsh
            for k in range(start_depth_index, -1, -1):
                if gradient[k] >= 0:
                    upward_depths[i, j] = depths_profile_grad[k]
                    break

# User input for latitude and longitude
user_lat = float(input("Enter latitude (5 to 30): "))
user_lon = float(input("Enter longitude (90 to 120): "))

# Find the index closest to the user's latitude and longitude
lat_idx = np.abs(lat_array - user_lat).argmin()
lon_idx = np.abs(lon_array - user_lon).argmin()

# Extract the smoothened Vsh profile for the user's location within 0-300 km range
smooth_vsh_profile_user = smooth_vsh_subset_grad_range[:, lat_idx, lon_idx]

# Find the minimum Vsh and its corresponding depth for the user location in the 40-300 km range
min_vsh_val_user = np.min(smooth_vsh_subset_vsh_range[:, lat_idx, lon_idx])
min_vsh_idx_user = np.where(smooth_vsh_subset_vsh_range[:, lat_idx, lon_idx] == min_vsh_val_user)[0][0]
min_vsh_depth_user = dp_array[dp_indices_vsh[min_vsh_idx_user]]

# Calculate the gradient of the smoothened Vsh profile (0-300 km range)
gradient_user = np.gradient(smooth_vsh_profile_user, dp_array[dp_indices_grad])

# Find the depth where the gradient changes from negative to neutral/positive (above the minimum Vsh point)
gradient_change_indices = np.where(gradient_user[:min_vsh_idx_user] >= 0)[0]

# Check if we found any indices
if gradient_change_indices.size > 0:
    gradient_change_idx = gradient_change_indices[-1]  # Get the last index where the gradient is neutral or positive
    gradient_change_depth = dp_array[dp_indices_grad[gradient_change_idx]]
else:
    gradient_change_depth = np.nan  # No gradient change found
    print(f"No gradient change found above the minimum Vsh at lat={user_lat}, lon={user_lon}.")  # Debug message

# Plotting the Vsh profile
plt.figure(figsize=(6, 8))
plt.plot(smooth_vsh_profile_user, dp_array[dp_indices_grad], label='Smoothened Vsh', color='blue')

# Mark the minimum Vsh point
plt.scatter(min_vsh_val_user, min_vsh_depth_user, color='blue', s=100, label=f'Min Vsh({min_vsh_depth_user:.2f} km)', marker='v')

# Mark the gradient change point if it exists
if not np.isnan(gradient_change_depth):
    plt.scatter(smooth_vsh_profile_user[gradient_change_idx], gradient_change_depth, color='orange', s=100, label=f'Gradient Change({gradient_change_depth:.2f} km)', marker='^')
    
    halfway_depth = (min_vsh_depth_user + gradient_change_depth) / 2
    plt.axhline(y=halfway_depth, color='red', linestyle='--', label=f'LAB Depth({halfway_depth} km)', linewidth=1.5) # Plotting LAB on that profile

plt.xlabel('Smoothened Vsh (km/s)')
plt.ylabel('Depth (km)')
plt.title(f'{user_lat}N {user_lon}E : Smoothened Vsh Profile (Moving average[window={win}])')
plt.gca().invert_yaxis()
plt.grid(True)
plt.legend()

odir=f"/media/rupak/Rupak_4TB/BACKUPS/MANUSCRIPTS_Rupak/SE_ASIA_MANU_DRAFTS/MERGED/Manu_figs_organised/text_fig/GitHub_upload/LAB/files"
plt.savefig(os.path.join(odir,f"Smoothened_VSH_profile_MA_marked_{user_lat}N_{user_lon}E.png"))
#plt.savefig(f"Smoothened_VSH_profile_MA_marked_{user_lat}N_{user_lon}E.pdf")
plt.show()
```

# Creating the file that holds the Lat, long, LAB depth, minVsh, min Vsh depth and gradient change depth

```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import matplotlib.ticker as ticker
import os

# Load the NetCDF data
ncf = f"/media/rupak/Rupak_4TB/BACKUPS/MANUSCRIPTS_Rupak/SE_ASIA_MANU_DRAFTS/MERGED/Manu_figs_organised/text_fig/GitHub_upload/model/FWEA23.r0.0-n4.nc"
nf = Dataset(ncf, 'r')
dp = nf.variables['depth'][:]
lon = nf.variables['longitude'][:]
lat = nf.variables['latitude'][:]
vsh = nf.variables['vsh'][:]

# Arrays
dp_array = np.array(dp)
lon_array = np.array(lon)
lat_array = np.array(lat)
vsh_array = np.array(vsh)

# Smooth the Vsh data
def apply_moving_average_vsh(vsh_data, win):
    dp_dim, lat_dim, lon_dim = vsh_data.shape
    smoothed_vsh = np.empty((dp_dim - win + 1, lat_dim, lon_dim))

    for i in range(vsh.shape[1]):
        for j in range(vsh.shape[2]):
            depth_profile = vsh_data[:, i, j]
            smoothed_profile = moving_average(depth_profile, win)
            smoothed_vsh[:, i, j] = smoothed_profile

    return smoothed_vsh

def moving_average(vsh, win):
    return np.convolve(vsh, np.ones(win) / win, mode='valid')

# Apply smoothing
win = 7  # Window size for smoothing
smoothed_vsh_array = apply_moving_average_vsh(vsh_array, win)

# Define depth ranges
dp_min_vsh_range = 40  # Minimum depth for finding Vsh_min
dp_max_vsh_range = 300  # Maximum depth for finding Vsh_min
dp_min_grad_range = 0  # Minimum depth for gradient calculation
dp_max_grad_range = 300  # Maximum depth for gradient calculation

# Extract depth indices within each range
dp_indices_vsh = np.where((dp_array >= dp_min_vsh_range) & (dp_array <= dp_max_vsh_range))[0]
dp_indices_grad = np.where((dp_array >= dp_min_grad_range) & (dp_array <= dp_max_grad_range))[0]

# Extract smoothened Vsh within the defined depth ranges
smooth_vsh_subset_vsh_range = smoothed_vsh_array[dp_indices_vsh, :, :]
smooth_vsh_subset_grad_range = smoothed_vsh_array[dp_indices_grad, :, :]

# User-defined lat-long bounds
lat_min, lat_max = 5, 30
lon_min, lon_max = 90, 120
lat_indices = np.where((lat_array >= lat_min) & (lat_array <= lat_max))[0]
lon_indices = np.where((lon_array >= lon_min) & (lon_array <= lon_max))[0]

# File to save results
#output_file = "LAB_depths_results.txt"
odir=f"/media/rupak/Rupak_4TB/BACKUPS/MANUSCRIPTS_Rupak/SE_ASIA_MANU_DRAFTS/MERGED/Manu_figs_organised/text_fig/GitHub_upload/LAB/files"
output_file=os.path.join(odir,f"Lat_lon_LABDEPTH_moving_average_window_{win}.txt")

with open(output_file, "w") as f:
    # Write headers
    ##f.write("Latitude Longitude LAB_Depth Min_Vsh_Value Min_Vsh_Depth Gradient_Change_Depth\n")

    # Loop over specified lat-long range
    for lat_idx in lat_indices:
        for lon_idx in lon_indices:
            # Get the smoothened Vsh profile within 40-300 km range for min Vsh
            smooth_vsh_profile_vsh_range = smooth_vsh_subset_vsh_range[:, lat_idx, lon_idx]

            # Find the minimum Vsh and its depth
            min_vsh_val = np.min(smooth_vsh_profile_vsh_range)
            min_vsh_idx = np.where(smooth_vsh_profile_vsh_range == min_vsh_val)[0][0]
            min_vsh_depth = dp_array[dp_indices_vsh[min_vsh_idx]]

            # Get the smoothened Vsh profile within 0-300 km range for gradient calculation
            smooth_vsh_profile_grad_range = smooth_vsh_subset_grad_range[:, lat_idx, lon_idx]
            depths_profile_grad = dp_array[dp_indices_grad]

            # Calculate the gradient within the 0-300 km range
            gradient_profile = np.gradient(smooth_vsh_profile_grad_range, depths_profile_grad)

            # Find the depth where gradient changes from negative to neutral/positive above min Vsh depth
            gradient_change_indices = np.where(gradient_profile[:min_vsh_idx] >= 0)[0]
            if gradient_change_indices.size > 0:
                gradient_change_idx = gradient_change_indices[-1]
                gradient_change_depth = depths_profile_grad[gradient_change_idx]
            else:
                gradient_change_depth = np.nan  # No gradient change found

            # Calculate the LAB depth (midpoint between min Vsh depth and gradient change depth)
            if not np.isnan(gradient_change_depth):
                lab_depth = (min_vsh_depth + gradient_change_depth) / 2
            else:
                lab_depth = np.nan  # No valid LAB depth if gradient change depth not found

            # Write to file
            f.write(f"{lat_array[lat_idx]} {lon_array[lon_idx]} {lab_depth:.2f} "
                    f"{min_vsh_val} {min_vsh_depth:.2f} {gradient_change_depth:.2f}\n")

print(f"Results saved to {output_file}")
```
